```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('vgsales.csv')
df
```

### Which company is the most common video game publisher?

```python

sort_common_video = df.groupby('Publisher').count()
most_common_video = sort_common_video.sort_values(by='Global_Sales', ascending=False).head(1).index[0]
assert most_common_video == 'Electronic Arts'
```

### What’s the most common platform?

```python
sort_commom_platform = df.groupby('Platform').count()
most_common_platform = sort_commom_platform.sort_values(by='Global_Sales', ascending=False).head(1).index[0]
assert most_common_platform == 'DS'
```

### What about the most common genre?


```python
sort_commom_genre = df.groupby('Genre').count()
most_common_genre = sort_commom_genre.sort_values(by='Global_Sales', ascending=False).head(1).index[0]
assert most_common_genre == 'Action'
```

### What are the top 20 highest grossing games?


```python
top_20 = df[["Name",'Global_Sales']]
sortGlobal = top_20.sort_values(by='Global_Sales', ascending=False)
highest_grossing_games = sortGlobal.head(20)
# for i in range(len(highest_grossing_games)):
#     print(highest_grossing_games[i])
highest_grossing_games
```

### For North American video game sales, what’s the median?

```python
Nort_American = float(df[['NA_Sales']].median())
assert (Nort_American) == 0.08
```

#### For the top-selling game of all time, how many standard deviations above/below the mean are its sales for North America?


```python
mean_NA=df['NA_Sales'].mean()
data_row=df['NA_Sales'].max()
std_NA=df['NA_Sales'].std()
deviations_from_mean= ((data_row-mean_NA)/std_NA)
assert float(deviations_from_mean)== 50.47898767479108
```
